```python
import numpy as np
import math
x = [[0.1, 0.7, 0.2, 0.0],[0.0, 0.5, 0.5, 0.0],[0.2, 0.2, 0.2, 0.4],[0.8, 0.1, 0.0, 0.1],[0.3, 0.0, 0.4, 0.3],[0.0, 0.4, 0.0, 0.6],[0.5, 0.0, 0.4, 0.1],[0.6, 0.3, 0.0, 0.1],[0.0, 0.5, 0.1, 0.4],[0.1, 0.6, 0.0, 0.3],[0.3, 0.2, 0.1, 0.4],[0.1, 0.5, 0.4, 0.0],[0.2, 0.0, 0.2, 0.6],[0.2, 0.6, 0.1, 0.1],[0.1, 0.7, 0.1, 0.1],[0.2, 0.4, 0.2, 0.2]]
R = np.zeros([16,16])
R_prime = np.zeros([16,16])
R_prime_1 = np.zeros([16,16])
R_prime_2 = np.zeros([16,16])
R_prime_3 = np.zeros([16,16])
R_prime_4 = np.zeros([16,16])
R_alpha_4 = np.zeros([16,16])
R_alpha_6 = np.zeros([16,16])
R_alpha_8 = np.zeros([16,16])
dot = 0
mag_i = 0
mag_j = 0
for i in range(16):
  for j in range(16):
    for k in range(4):
      dot += x[i][k]*x[j][k]
      mag_i += x[i][k]*x[i][k]
      mag_j += x[j][k]*x[j][k]
    R[i][j] = abs(dot)/(math.sqrt(mag_i*mag_j))
    dot = 0
    mag_i = 0
    mag_j = 0


```

```python
import skfuzzy
#R_prime = R
#R_prime[2][4] = R[2][4]
R_prime = skfuzzy.fuzzymath.maxmin_composition(R, R)
#R_prime = skfuzzy.fuzzymath.maxmin_composition(R_prime, R_prime)
#if (R_prime.all() == R.all()):
#  print("budududu")
#else:
#  print("badadadada")
#R_prime == R
```

```python
R_prime_1 = skfuzzy.fuzzymath.maxmin_composition(R_prime, R_prime)
R_prime_2 = skfuzzy.fuzzymath.maxmin_composition(R_prime_1, R_prime_1)
R_prime_3 = skfuzzy.fuzzymath.maxmin_composition(R_prime_2, R_prime_2)
#R_prime_2 == R_prime_3
```

```python
for i in range(16):
  for j in range(16):
    if (R_prime_3[i][j]>=0.4):
      R_alpha_4[i][j] = 1
```
